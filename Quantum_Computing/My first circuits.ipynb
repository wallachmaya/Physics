Below is an example of a simple quantum circuit (used on quantum computers) next to a classical circuit (used on binary or regular computers).

```python
from qiskit import(
  QuantumCircuit,
  execute)
from qiskit.visualization import plot_histogram
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
#In the parenthesis the first number represents how many Quantum circuts will be used while the next number reprents how
#many classical circuts are used
circut=QuantumCircuit(1,1)
circut.x(0)
circut.draw()
```

In this cell I used the statevector_simulator to do the same thing as above.

```python
import numpy as np
from qiskit import *
from qiskit.tools.visualization import plot_bloch_multivector
#here we make a circut with one classical bit and one Quantum bit
circut=QuantumCircuit(3)
#This makes whatever number inside parenthesis in state one
#If you do multiple of these commands then 
circut.x(0)
simulator=Aer.get_backend("statevector_simulator")
result = execute(circut, backend=simulator).result()
statevector=result.get_statevector()
print(statevector)
plot_bloch_multivector(statevector)
```

I found the code to make more appealing drawings.

```python
circut.draw(output='mpl')
```

```python
import numpy as np
q = QuantumRegister(4)
qantCur = QuantumCircuit(q)
#ry represents the gate that is used to ditermine were the arrow points
qantCur.ry(np.pi/2,q)
qantCur.x(0)
qantCur.ry(np.pi/3,q)
qantCur.x(1)
qantCur.ry(np.pi/4,q)
qantCur.x(2)
qantCur.ry(np.pi/5,q)
qantCur.x(3)
simulator=Aer.get_backend("statevector_simulator")
result = execute(qantCur, backend=simulator).result()
statevector=result.get_statevector()
print(statevector)
plot_bloch_multivector(statevector)
```

This is what the circuit would look like

```python
qantCur.draw(output='mpl')
```

Here I replaced x with y and z.

```python
q = QuantumRegister(3)
qantCur = QuantumCircuit(q)
#ry represents the gate that is used to ditermine were the arrow points
qantCur.rx(np.pi/2,q)
qantCur.z(0)
qantCur.y(1)
qantCur.x(2)
simulator=Aer.get_backend("statevector_simulator")
result = execute(qantCur, backend=simulator).result()
statevector=result.get_statevector()
print(statevector)
plot_bloch_multivector(statevector)
```

```python
qantCur.draw(output="mpl")
```

Here I made a histogram ploting the Bell State of all 4 qubits from the 5th cell. Bell State is a maximally entangled quantum state of two qubits.

```python
from qiskit.tools.visualization import plot_histogram, plot_state_city
q = QuantumRegister(4)
qantCur = QuantumCircuit(q)
#ry represents the gate that is used to ditermine were the arrow points
qantCur.ry(np.pi/2,q)
qantCur.x(0)
qantCur.ry(np.pi/3,q)
qantCur.x(1)
qantCur.ry(np.pi/4,q)
qantCur.x(2)
qantCur.ry(np.pi/5,q)
qantCur.x(3)

# Select the StatevectorSimulator from the Aer provider
simulator = Aer.get_backend('statevector_simulator')

# Execute and get counts
result = execute(qantCur, simulator).result()
statevector = result.get_statevector(qantCur)
plot_state_city(statevector, title='Bell state')
```
